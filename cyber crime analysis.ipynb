# Bivariate Analysis

```python
# top 10 cities with highest crime rates

# Grouping data to get the total cybercrime count per city
city_crime_counts = df.groupby("City")["Total"].sum().reset_index()

# Sorting to get the top 10 cities with the highest cybercrime rates
top_cities = city_crime_counts.sort_values(by="Total", ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Total", y="City", data=top_cities, color='r')

# Adding titles and labels
plt.title("Top 10 Cities with Highest Cybercrime Rates", fontsize=14)
plt.xlabel("Total Cybercrime Cases", fontsize=12)
plt.ylabel("City", fontsize=12)

# Displaying the plot
plt.show()
```

ðŸ“Œ observation: 
1. vishakhapatnam has the highest crime rate 

```python
#heatmap to show the correlation between the different cybercrimes.

# Selecting only numeric columns
numeric_df = df.select_dtypes(include=["number"])

# Computing correlation matrix
corr_matrix = numeric_df.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".1f", linewidths=0.5)

# Adding a title
plt.title("Crime Correlation Heatmap", fontsize=14)
plt.show()
```

ðŸ“Œ observation:
#Red (Positive Correlation, closer to +1): If one crime increases, the other also increases.
#Blue (Negative Correlation, closer to -1): If one crime increases, the other decreases.
#White (Near 0 Correlation): No significant relationship between the crimes.

1. Terrorist Activities & Terrorist Recruitment (0.58)- Terrorism-related activities often involve recruitment efforts.
2. Sexual Exploitation & Psycho/Pervert Cases (0.62)- Online sexual crimes often involve psychologically disturbed offenders.
3. Personal Revenge & Spreading Piracy (-0.50)- People committing crimes for personal revenge are less likely involved in piracy-related offenses.

```python
# Pie chart for highest crime rate cities

# Assuming 'df' contains 'City' and 'Crime Rate' columns
df_sorted = df[['City', 'Crime Rate']].sort_values(by='Crime Rate', ascending=False)

# Selecting top 5 highest and lowest crime rate cities
top_cities = df_sorted.head(5)


# Merging both into one dataset
selected_cities = pd.concat([top_cities])

# Plotting the pie chart
plt.figure(figsize=(10, 6))
plt.pie(selected_cities['Crime Rate'], labels=selected_cities['City'], autopct='%1.1f%%', startangle=90)

# Title and display
plt.title('Cities with Highest Crime Rates')
plt.axis('equal')  # Ensures the pie chart remains circular
plt.show()
```

ðŸ“Œ observation: Vishakhapatnam has the highest crime rate than Varanasi than Allahabad
